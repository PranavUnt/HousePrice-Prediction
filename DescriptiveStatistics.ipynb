Bar plot of 25th, 50th, 75th percentiles for each feature

```python
p_a = [p1, p2, p3, p4, p5, p6, p7, p8, p9, p10, p11, p12, p13, p14, p15]
per = []
for i in range(5):
    per.append('25th percentile')
    per.append('50th percentile')
    per.append('75th percentile')

p = pd.DataFrame(p_a)
p.reset_index(drop=True, inplace=True)

groups = np.array_split(p, np.ceil(len(p) / 3))

num_groups = len(groups)
num_cols = min(num_groups, 3)
num_rows = np.ceil(num_groups / num_cols).astype(int)

fig, axs = plt.subplots(num_rows, num_cols, figsize=(15, 4 * num_rows))

if num_groups == 1:
    axs = np.array([axs])

for i, group in enumerate(groups):
    row = i // num_cols
    col = i % num_cols
    ax = axs[row, col]
    x = np.arange(3)
    bar_width = 0.3
    random_colors = np.random.rand(3, 3)
    bars = ax.bar(x, group.values.flatten(), bar_width, color=random_colors)

    ax.set_xlabel('Percentiles')
    ax.set_ylabel('Value')
    ax.set_title(features[i])
    ax.set_xticks(x)
    ax.set_xticklabels(per[i * 3: (i + 1) * 3])

    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height}', xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3), textcoords='offset points',
                    ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python

p = data_str['Town'].value_counts().sort_index()

groups = np.array_split(p, np.ceil(len(p) / 3))

num_groups = len(groups)
num_cols = min(num_groups, 3)
num_rows = np.ceil(num_groups / num_cols).astype(int)

fig, axs = plt.subplots(num_rows, num_cols, figsize=(15, 4 * num_rows), sharey=True)

if num_groups == 1:
    axs = [axs]

for i, group in enumerate(groups):
    row = i // num_cols
    col = i % num_cols
    ax = axs[row][col]
    bars = ax.bar(group.index, group.values, color='brown')
    ax.set_xlabel('Town')
    ax.set_ylabel('Frequency')
    ax.set_title(f'Group {i+1}')

    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height}', xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3), textcoords='offset points',
                    ha='center', va='bottom')

plt.tight_layout()

plt.show()
```
